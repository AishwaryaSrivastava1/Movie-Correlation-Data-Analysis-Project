# Movie Correlation Data Analysis Project

### This project finds which variables affect the gross revenues from movies. 
### Kaggle dataset, with 6820 rows and 15 columns and a size of 960 KB, is used here. 


```python
#Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 
pd.options.mode.chained_assignment = None

#Read the data
df=pd.read_csv('E:\Data Science\Data Analyst Projects\MovieIndustryKaggle\movies.csv',encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
#returns %age of missing data in each column
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing)) 
```

```python
df.dtypes
```

```python
df['budget']=df['budget'].astype(int)
df['gross']=df['gross'].astype(int)
df.dtypes
```

```python
df['year_correct']=df['released'].astype(str).str[:4]
df.head()
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
df.drop_duplicates()
```

```python
df['company'].value_counts()
```

### Checking which features will have a correlation with 'gross'
### Checking if Budget and Company have a high correlation with gross or not by building a scatterplot

```python
#Scatterplot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#Regression Plot for Budget vs Gross
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#let's start looking at correlation
df.corr()
```

```python
df.corr(method='pearson')
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

#### Conclusion: From all the above correlation matrices , we get high correlation between budget and gross

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized=df.copy()
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes  #gives random numerization

df_numerized
```

```python
correlation_matrix=df_numerized.corr(method='pearson') #change method to spearman,kendall to get different heatmaps
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```

### Conclusion:
      Votes and Budget have the highest correlation to gross earnings
      Company has low correlation with gross earnings
